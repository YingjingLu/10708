## 1.3.1. Example for Bayesian network
In the library, we represent a Bayesian network as a directed acyclic graph. The relationships between nodes are parameterized using Conditional Probability Distributions (CPD). Each node in the network is parameterized using $ P(node | Pa(node)) $ where $ Pa(node) $ represents the parents of node in the network.

Take following model as an example:

```python
from IPython.display import Image
Image('../figures/bn_graph_table.png')
```

In the library we define the network structure and the CPDs separately and then associate them with the structure. Here's an example for defining the above model:

```python
from cmulibpgm.models import BayesianModel
from cmulibpgm.factors.discrete import TabularCPD

# Defining the model structure. We can define the network by just passing a list of edges.
model = BayesianModel([('A', 'C'), ('B', 'C'), ('B', 'D'), ('C', 'E')])

# Defining individual CPDs.
cpd_a = TabularCPD(variable='A', variable_card=2, values=[[0.4, 0.6]])
cpd_b = TabularCPD(variable='B', variable_card=2, values=[[0.8, 0.2]])

# The representation of CPD in the library is a bit different than the CPD shown in the above picture. In the library the colums
# are the evidences and rows are the states of the variable. So the grade CPD is represented like this:
#
# +-----+-----+-----+-----+------+
# | A   | A_0 | A_0 | A_1 | A_1  |
# +-----+-----+-----+-----+------+
# | B   | B_0 | B_1 | B_0 | B_1  |
# +-----+-----+-----+-----+------+
# | C_0 | 0.2 | 0.1 | 0.3 | 0.8  |
# +-----+-----+-----+-----+------+
# | C_1 | 0.4 | 0.7 | 0.5 | 0.15 |
# +-----+-----+-----+-----+------+
# | C_2 | 0.4 | 0.2 | 0.2 | 0.05 |
 #+-----+-----+-----+-----+------+

cpd_c = TabularCPD(variable='C', variable_card=3, 
                   values=[[0.2, 0.1, 0.3, 0.8],
                           [0.4, 0.7, 0.5, 0.15],
                           [0.4, 0.2, 0.2, 0.05]],
                  evidence=['A', 'B'],
                  evidence_card=[2, 2])

cpd_d = TabularCPD(variable='D', variable_card=2, 
                   values=[[0.85, 0.1],
                           [0.15, 0.9]],
                   evidence=['B'],
                   evidence_card=[2])

cpd_e = TabularCPD(variable='E', variable_card=2,
                   values=[[0.95, 0.2],
                           [0.05, 0.8]],
                   evidence=['C'],
                   evidence_card=[2])

# Associating the CPDs with the network
model.add_cpds(cpd_a, cpd_b, cpd_c, cpd_d, cpd_e)
```

After defining the network, we have several methods associated with the network, such as checking whether the network structure and CPDs are valid, geting CPDs, geting cardinalities and so on. In the following are some examples how to do so. You may also refer to the source code for the methods we provide. In the future, we may require you to implement some methods yourself. 

```python
# check_model checks for the network structure and CPDs and verifies that the CPDs are correctly 
# defined and sum to 1.
model.check_model()
```

```python
# We can now call some methods on the BayesianModel object.
model.get_cpds()
```

```python
print(model.get_cpds('C'))
```

```python
model.get_cardinality('C')
```

You may replace one or more CPDs with new distributions. For example, we redefine the CPD for node D, and update the model. 

```python

cpd_d = TabularCPD(variable='D', variable_card=2, 
                   values=[[0.8, 0.3],
                           [0.2, 0.7]],
                   evidence=['B'],
                   evidence_card=[2])

# Update the CPD for node D
model.add_cpds(cpd_d)

# check_model checks for the network structure and CPDs and verifies that the CPDs are correctly 
# defined and sum to 1.
model.check_model()
```

## 1.3.2. An exercise for representing Bayesian Network

As you have seen how we could represent a Bayesian Network in the library, there is an exercise to work with on the library yourself. Below is a Bayesian network, please answer the following questions:
1. Represent the network with the library, and check whether it is a valid network. If not, leave the error information, and make corrections of the incorrect CPD(s) in another cell to get a valid network, then answer the follow questions.   
2. Show the CPD of node D.
3. Get the cardinality of node E. 

```python
Image('../figures/bn_graph_table_1.png')
```

```python
from cmulibpgm.models import BayesianModel
from cmulibpgm.factors.discrete import TabularCPD

# Defining the model structure. We can define the network by just passing a list of edges.
model = BayesianModel([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('B', 'F'), ('C', 'E'), ('D', 'F')])

# Defining individual CPDs.
cpd_a = TabularCPD(variable='A', variable_card=2, values=[[0.3, 0.7]])
cpd_b = TabularCPD(variable='B', variable_card=2, 
                   values=[[0.5, 0.6], 
                           [0.5, 0.4]], 
                   evidence=['A'],
                   evidence_card=[2])

cpd_c = TabularCPD(variable='C', variable_card=2, 
                   values=[[0.1, 0.7],
                           [0.9, 0.3]],
                  evidence=['A'],
                  evidence_card=[2])

cpd_d = TabularCPD(variable='D', variable_card=3, 
                   values=[[0.25, 0.7, 0.2, 0.6],
                           [0.35, 0.2, 0.6, 0.25],
                           [0.4,  0.1, 0.2, 0.15]],
                   evidence=['B', 'C'],
                   evidence_card=[2, 2])

cpd_e = TabularCPD(variable='E', variable_card=4,
                   values=[[0.19, 0.36],
                           [0.35, 0.23],
                           [0.31, 0.25], 
                           [0.15, 0.14]],
                   evidence=['C'],
                   evidence_card=[2])

cpd_f = TabularCPD(variable='F', variable_card=2,
                   values=[[0.1, 0.3, 0.45, 0.7, 0.34, 0.85],
                           [0.9, 0.7, 0.55, 0.3, 0.66, 0.15 ]],
                   evidence=['B', 'D'],
                   evidence_card=[2, 3])

# Associating the CPDs with the network
model.add_cpds(cpd_a, cpd_b, cpd_c, cpd_d, cpd_e, cpd_f)
```

```python
# check_model checks for the network structure and CPDs and verifies that the CPDs are correctly 
# defined and sum to 1.
model.check_model()
```

Clearly, the CPD for node E is invalid, because the conditional probability not sum to 1. Here is an example of correction:


```python
cpd_e = TabularCPD(variable='E', variable_card=4,
                   values=[[0.19, 0.36],
                           [0.35, 0.23],
                           [0.31, 0.27], 
                           [0.15, 0.14]],
                   evidence=['C'],
                   evidence_card=[2])
model.add_cpds(cpd_e)
```

```python
model.check_model()
```

```python
print(model.get_cpds('D'))
```

```python
model.get_cardinality('E')
```
